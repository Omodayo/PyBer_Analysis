```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plot
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
# ADD ERROR BARS
```

```python
# Add error bars a LINE CHARTS using MATLAB method.
# Add a cap to the error bars that is equal to a. scalar value.
plot.subplots(figsize=(12, 6))
plot.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)

plot.show()
```

```python
# Add error bars to a LINE CHARTS using object-oriented method.
# Add a cap to the error bars that is equal to a. scalar value.
fig, ax = plot.subplots(figsize=(12, 6))
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)

plot.show()
```

```python
# Add error bars a BAR CHARTS using MATLAB method.
# Add a cap to the error bars that is equal to a. scalar value.
plot.subplots(figsize=(12, 6))
plot.bar(x_axis, y_axis, yerr=stdev, capsize=3)

plot.show()
```

```python
# Add error bars to a BAR CHARTS using object-oriented method.
# Add a cap to the error bars that is equal to a. scalar value.
fig, ax = plot.subplots(figsize=(12, 6))
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)

plot.show()
```

```python
# CHANGING THE MAJOR TICKS
```

```python
# To adjust the major x-ticks on a horizontal bar chartusing MATLAB method, we use xticks() function
import numpy as np
from matplotlib.ticker import MultipleLocator

plot.subplots(figsize=(15, 12))
plot.barh(x_axis, y_axis)
plot.xticks(np.arange(0, 51, step=5.0))
plot.gca().invert_yaxis

#Add minor ticks between the major ticks on the x-axis at an increament of 1.
#ax.xaxis.set_minor_locator(MultipleLocator(1))

plot.show()
```

```python
# To adjust the major x-ticks on a horizontal bar chartusing object-oriented approach, we use xticks() function
import numpy as np
from matplotlib.ticker import MultipleLocator

fig, ax = plot.subplots(figsize=(15, 12))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 50, step=5.0))
ax.invert_yaxis

#Add minor ticks between the major ticks on the x-axis at an increament of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))

plot.show()
```
