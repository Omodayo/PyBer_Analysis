```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
# file_to_read = 'Resources/PyBer_ride_data.csv'
file_to_load = os.path.join('Resources', 'PyBer_ride_data.csv')
ride_data_df = pd.read_csv(file_to_load)
ride_data_df
```

```python
ride_data_df.plot(x='Month', y='Avg. Fare ($USD)', figsize=(12, 8))

plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(ride_data_df))
tick_locations = [value for value in x_axis]

# Plot the data
ride_data_df.plot(x='Month', y='Avg. Fare ($USD)', figsize=(12, 8))
plt.xticks(tick_locations, ride_data_df['Month'])

plt.show()
```

```python
ride_data_df.plot.bar(x='Month', y='Avg. Fare ($USD)', figsize=(12, 8))

plt.show()
```

```python
# Annotate the bar chart
  # Add error bar.
    # Add caps to the error bars.
    # Change the color of the bars to sky blue.
    # Rotate the labels on the x-axis to horizontal
    # Set the y-axis increment to every $5.
import statistics

stdev = statistics.stdev(ride_data_df['Avg. Fare ($USD)'])
ride_data_df.plot.bar(x='Month', y='Avg. Fare ($USD)', yerr=stdev, capsize=3,
                      figsize=(12, 7), color='skyblue', align='center')

plt.yticks(np.arange(0, 50, step=5.0))
plt.xticks(rotation=360, ha='center')

plt.show()
```
